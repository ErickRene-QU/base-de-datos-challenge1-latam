### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_t1 = tienda['Precio'].sum()
ingreso_t2 = tienda2['Precio'].sum()
ingreso_t3 = tienda3['Precio'].sum()
ingreso_t4 = tienda4['Precio'].sum()

print(f"Tienda 1: ${ingreso_t1:,.2f}")
print(f"Tienda 2: ${ingreso_t2:,.2f}")
print(f"Tienda 3: ${ingreso_t3:,.2f}")
print(f"Tienda 4: ${ingreso_t4:,.2f}")
```

# 2. Ventas por categoría

```python
print("--- Tienda 1 ---\n", tienda['Categoría del Producto'].value_counts())
print("\n--- Tienda 2 ---\n", tienda2['Categoría del Producto'].value_counts())
print("\n--- Tienda 3 ---\n", tienda3['Categoría del Producto'].value_counts())
print("\n--- Tienda 4 ---\n", tienda4['Categoría del Producto'].value_counts())
```

# 3. Calificación promedio de la tienda


```python
calif_t1 = tienda['Calificación'].mean()
calif_t2 = tienda2['Calificación'].mean()
calif_t3 = tienda3['Calificación'].mean()
calif_t4 = tienda4['Calificación'].mean()

print(f"Tienda 1: {calif_t1:.2f} estrellas")
print(f"Tienda 2: {calif_t2:.2f} estrellas")
print(f"Tienda 3: {calif_t3:.2f} estrellas")
print(f"Tienda 4: {calif_t4:.2f} estrellas")
```

# 4. Productos más y menos vendidos

```python
data_tiendas = [tienda, tienda2, tienda3, tienda4]

for i, t in enumerate(data_tiendas, 1):
    ventas = t['Producto'].value_counts()
    print(f"\n--- TIENDA {i} ---")
    print("Top 3 Más Vendidos:\n", ventas.head(3))
    print("\nTop 3 Menos Vendidos:\n", ventas.tail(3))
```

# 5. Envío promedio por tienda

```python
envio_t1 = tienda['Costo de envío'].mean()
envio_t2 = tienda2['Costo de envío'].mean()
envio_t3 = tienda3['Costo de envío'].mean()
envio_t4 = tienda4['Costo de envío'].mean()

print(f"Tienda 1: ${envio_t1:,.2f}")
print(f"Tienda 2: ${envio_t2:,.2f}")
print(f"Tienda 3: ${envio_t3:,.2f}")
print(f"Tienda 4: ${envio_t4:,.2f}")
```

Visualizaciones y Reto Extra Geográfico

```python
import matplotlib.pyplot as plt

# 1. Gráfico de Barras: Comparación de Ingresos
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_t1, ingreso_t2, ingreso_t3, ingreso_t4]

plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos, color=['#5cb85c', '#5bc0de', '#f0ad4e', '#d9534f'])
plt.title('Ingreso Total por Tienda', fontsize=14, fontweight='bold')
plt.ylabel('Facturación')
plt.show()

# 2. Gráfico de Pastel: Distribución de Categorías (Ejemplo Tienda 1)
categorias_t1 = tienda['Categoría del Producto'].value_counts()
# Paleta de colores elegante y a prueba de errores:
colores = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#c2c2f0', '#ffb3e6', '#c4e17f', '#f2c2e0']

plt.figure(figsize=(8, 8))
plt.pie(categorias_t1, labels=categorias_t1.index, autopct='%1.1f%%', startangle=90, colors=colores)
plt.title('Distribución de Categorías - Tienda 1', fontsize=14, fontweight='bold')
plt.show()

# 3. Reto Extra Geográfico: Gráfico de Dispersión (Mapa de ventas Tienda 1)
plt.figure(figsize=(8, 6))
plt.scatter(tienda['lon'], tienda['lat'], alpha=0.5, color='purple')
plt.title('Distribución Geográfica de Compras - Tienda 1', fontsize=14, fontweight='bold')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
